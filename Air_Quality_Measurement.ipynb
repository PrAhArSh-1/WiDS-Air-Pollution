## Exploratory Data Analysis


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# importing the dataset
df_1=pd.read_excel('site_139620221221110426.xlsx',skiprows=16,nrows=1080)
df_1.head()
```

```python
df_1.columns
```

```python
df_1.tail()
```

```python
df_1.describe()
```

```python
df_1.shape
```

```python
df_1.rename(columns = {'PM2.5':'PM25'}, inplace = True)
```

# Number of rows with NO values

```python
df_1[df_1['PM2.5']=="None"].value_counts()
```

```python
df_1.isnull().sum()
```

```python
df_1.replace(to_replace="None",value=0.00,inplace=True)
df_1.head()
```

# Pandas Dataframe visualization


```python
df_1["From Date"]=pd.to_datetime(df_1['From Date'],format="%d-%m-%Y %H:%M")
df_1.head()
```

```python
type(df_1['From Date'][1])
```

# Creating a new column 'DAY'

```python
df_1['DAY']=df_1.index+1
df_1.head()
```

# Visualization

```python
sns.pairplot(df_1)
```

```python
df_1.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_1.corr(),annot=True)
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(df_1.DAY,df_1.PM25,color='magenta')
plt.title("The Observation Of the Day VS PM2.5")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(df_1.DAY,df_1.PM10,color='blue')
plt.title("The Observation Of the Day VS PM10")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(df_1.DAY,df_1.SO2,color='yellow')
plt.title("The Observation Of the Day VS SO2")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(df_1.DAY,df_1.NO2,color='green')
plt.title("The Observation Of the Day VS SO2")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(df_1.DAY,df_1.NH3,color='orange')
plt.title("The Observation Of the Day VS NH3")
plt.grid()
plt.show()
```
